```python
# Standadizing the room values
rooms_std = []

for val in X.Room:
    if val == 0:
        rooms_std.append(0)
    else:
        rooms_std.append(int(val) / int(len(X.Room)))

X.Room = rooms_std
```

```python
# Standadizing the luxury values
Service = []
Court = []
Mall = []
Spa = []
Deck = []

headers = ['FoodCourt', 'ShoppingMall']

for val in X.RoomService:
    if val == 0:
        Service.append(0)
    else:
        Service.append(1 / int(val))

for val in X.FoodCourt:
    if val == 0:
        Court.append(0)
    else:
        Court.append(1 / int(val))

for val in X.ShoppingMall:
    if val == 0:
        Mall.append(0)
    else:
        Mall.append(1 / int(val))

for val in X.Spa:
    if val == 0:
        Spa.append(0)
    else:
        Spa.append(1 / int(val))

for val in X.VRDeck:
    if val == 0:
        Deck.append(0)
    else:
        Deck.append(1 / int(val))

X.FoodCourt = Court
X.ShoppingMall = Mall
X.Spa = Spa
X.VRDeck = Deck
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

y_test = y_test.astype(int)
y_train = y_train.astype(int)
```

```python
def getmae(xtrain, xval, ytrain, yval, iter, layer):
    model = MLPRegressor(random_state=1, max_iter=1000, hidden_layer_sizes=500).fit(xtrain, ytrain)
    prediction = model.predict(xval)
    prediction = prediction.round()
    mae = mean_absolute_error(yval, prediction)
    return mae
```

```python
test_nodes = [50, 100, 200, 300, 500]
test_iter = [100, 200, 300, 400, 500]

best_fit = [1, 0, 0]

for vali in test_nodes:
    for valj in test_iter:
        pred = getmae(X_train, X_test, y_train, y_test, valj, vali)
        if pred < best_fit[0]:
            best_fit[1] = valj
            best_fit[2] = vali
            best_fit[0] = pred
print(best_fit)
```
